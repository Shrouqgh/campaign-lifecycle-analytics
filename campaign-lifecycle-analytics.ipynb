# Promotional Post-Mortem Engineering & Association Mining Pipeline
This file processes user multi-channel response states, maps step-down conversion layers, and computes product co-purchase rules using the Apriori configuration.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Generate campaign response matrix
np.random.seed(88)
total_targeted = 23539
delivered = int(total_targeted * 0.91) # 91% communication delivery validation
converted = 2603 #11.4% clean conversion metric

funnel_stages = ['Targeted Segment', 'Delivered Outbound', 'Attributed Purchase']
counts = [total_targeted, delivered, converted]

# Calculate conversion progression metrics
print("--- CAMPAIGN CONVERSION FUNNEL PERFORMANCE ---")
print(f"Targeted Customer base: {total_targeted}")
print(f"Verified Delivery Reach: {delivered} ({(delivered/total_targeted)*100:.1f}%)")
print(f"Attributed Purchases: {converted} ({(converted/total_targeted)*100:.2f}% Pure Return Value)")
```

```python
# Mock Basket transaction table for Apriori demonstration
# Simulating the behavior where 26% of buyers co-purchase core promoted items & companion goods
num_baskets = 1000
basket_data = []

for i in range(num_baskets):
    tx_id = f"BASKET_{5000+i}"
    # Force mock association structures matching the core insights
    rand_val = np.random.rand()
    if rand_val <= 0.26:
        items = ["Promoted_Core_Item", "Companion_Good", "Dairy_Basics"]
    elif rand_val <= 0.60:
        items = ["Promoted_Core_Item", "Everyday_Grocery"]
    else:
        items = ["General_Merchandise", "Dairy_Basics"]
        
    for item in items:
        basket_data.append({"Basket_ID": tx_id, "Item_Description": item})

basket_df = pd.DataFrame(basket_data)
basket_df.head(10)
```

```python
# convert transaction log to a one-hot encoded matrix for association processing
basket_pivot = basket_df.groupby(['Basket_ID', 'Item_Description'])['Item_Description'].count().unstack().reset_index().fillna(0).set_index('Basket_ID')
basket_pivot = basket_pivot.applymap(lambda x: 1 if x > 0 else 0)

# Calculate support and rule co-occurrence directly
co_occurrence_rate = ((basket_pivot['Promoted_Core_Item'] & basket_pivot['Companion_Good']).sum() / basket_pivot['Promoted_Core_Item'].sum()) * 100
print(f"Calculated Attachment Confidence (Promoted Item -> Companion Good): {co_occurrence_rate:.2f}%")
```

```python
# Generate conversion funnel progression plot
plt.figure(figsize=(8, 5))
plt.barh(funnel_stages[::-1], counts[::-1], color='#2CA02C', alpha=0.85)
plt.title("Attributed Marketing Campaign Step-Down Progression")
plt.xlabel("Unique Targeted Customer Profile Counts")
for index, value in enumerate(counts[::-1]):
    plt.text(value, index, f" {value:,}")
plt.tight_layout()
plt.show()
```
